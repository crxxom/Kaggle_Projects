# Import Dataset

```python
import pandas as pd
train_df = pd.read_csv('/kaggle/input/titanic/train.csv')
test_df = pd.read_csv('/kaggle/input/titanic/test.csv')
```

# Pycaret

```python
!pip install pycaret
```

```python
from pycaret.classification import setup, compare_models, evaluate_model,predict_model, save_model, load_model
```

```python
s = setup(train_df, target='Survived')
best_model = compare_models()
```

```python
evaluate_model(best_model)
```

```python
prediction = predict_model(best_model, data=test_df)
prediction
```

```python
submission = pd.DataFrame({'PassengerId': test_df.PassengerId.values,
                          'Survived': prediction.prediction_label.values})
submission.to_csv('Titanic_pycaret3.csv',index=False)
```

A submission score of 0.78468, you can improve the score significantly furthur by fine tuning the model seperately, changing the setup or using better feature selection techniques, feel free to copy the notebook and mess around with different settings yourself for further learning purpose.
